```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
#for plots to be in JN console
```

```python
#create a random dataframe
np.random.seed(20)
df1=pd.DataFrame(np.random.rand(1000,1), index=pd.date_range('2022-06-07', periods=1000), columns=['Data'])
df1.head()
```

```python
#just looking at data won't give us any explanations, so we will use visualization
df1.plot()

#lables on x-axis(here, Date) and feature on y-axis(here, Data)
```

```python
#styling
df1.plot(style='b*', legend=True, label="Madhu")
#legend=label appearing left-down
#label is not working in this case?
```

```python
pd.DataFrame(df1.Data.cumsum().head())
```

```python
#community sum('Data') plot
df1.Data.cumsum().plot()
```

```python
df1.head()
```

```python
#let's make some new columns in df1
df1['cumsum']=df1['Data'].cumsum()
df1['twice']=df1['Data'].apply(lambda x : 1/x)
```

```python
df1.head()
```

```python
#now plot
df1.plot(x=None, y=['twice'])
```

### Histogram

**histogram--> x represents range of data and y represents frequency of data in that range**

```python
#iris dataset
df2.head()
```

```python
df2.plot(kind='hist')
#overlapping data, so we can stack or plot separately
```

```python
df2.plot(kind='hist', stacked=True)
#yes we can now see overlapped data
```

```python
df2.plot(kind='hist', stacked=True, bins=100)
#bins controls width
```

```python
df2.describe()
```

```python
df2.petal_length.plot.hist()
#separate histogram, means say in range(1, appox 2) of sepal_length we have approx 35 data

#with the help of histogram we can draw normal distribution curve, which will be of great use in coming lectures
#we can see if our data is evenly distributed or not(in this case, no)
```

```python
df2.plot(figsize=(10,10), kind='hist', orientation='horizontal', stacked=True, bins=100, title='Histogram for iris dataset')
```

```python
#separate grphs
df2.hist(figsize=(10,10), color='#003333', alpha=0.7)
#color-->html color picker
#alpha=intensity of color
```

### scatter plot-->relation b/w two features

```python
df2.head()
```

```python
#compare sepal and petal length
#here we have to specify x and y
df2.plot(kind='scatter', x='sepal_length', y='petal_length')
```

```python
df2.plot(kind='scatter', x='sepal_length', y='sepal_width', figsize=(10,5))
#color of sepal length depends upon sepal width
```

```python
df2.plot(kind='scatter', x='petal_width', y='petal_length', c='sepal_length')
#looks like linear relation
#color depnds on sepal_length

#sepal_length	sepal_width	petal_length	petal_width	
#5.1	3.5	1.4	0.2	

#to create third dependency we plot such graphs
```

```python
#beautifications
df2.plot(kind='scatter', x='sepal_length', y='sepal_width', c='petal_length', label='Madhu', s=50)
#s-->size
```

```python
#hexabin --> shape..
df2.plot(kind='hexbin', x='sepal_length', y='sepal_width', label='Madhu', gridsize=10)
```
