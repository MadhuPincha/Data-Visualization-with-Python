##### We saw that earlier we have plotted 2 curves in the same plot but sometimes we have to plot graphs separately. So for that we'll use subplots in matplotlib

```python
# create a plot figure
plt.figure(figsize=(18,6))  

# lets plot a graph in 2 rows and one column-
# create the first of two panels and set current axis
#          (rows, columns, panel number)
plt.subplot(2,    1,       1) 
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

Alternatively you can use the below method

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
