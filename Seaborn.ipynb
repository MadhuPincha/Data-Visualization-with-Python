```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=sns.load_dataset('tips')
```

```python
df1.head()
```

### relplot

```python
sns.relplot(data=df1, x='tip', y='total_bill')
#kind=scatter by default
```

```python
sns.relplot(data=df1, x='tip', y='total_bill', hue='smoker', size='size', style='sex', height=10, aspect=1, legend='brief')
#hue-->color based on smoker categories
#style-->shapes
#size-->size of bubbles
#all these parameters are made so as to group features
```

```python
sns.relplot(data=df1, x='tip', y='total_bill', kind='line', hue='day', height=5, aspect=2)
#line
```

```python
#random dataframe
df2=pd.DataFrame(np.random.rand(300,2), columns=['a', 'b'])
```

```python
df2.head()
```

```python
sns.relplot(data=df2, x='a', y='b', kind='line')
```

```python
sns.relplot(data=df2, x='a', y='b', kind='line', sort=True)
```

### catplot()
categorical vs numerical

```python
#tip
df1.head()
```

```python
sns.catplot(data=df1, x='day', y='tip')
#kind-->"strip"(default), "swarm", "box", "violin", "boxen", "point", "bar", or "count".
```

```python
sns.catplot(data=df1, x='smoker', y='total_bill', kind='box',hue='sex', aspect=2)
#kind-->"strip"(default), "swarm", "box", "violin", "boxen", "point", "bar", or "count".
#so female usually pays less amount
```

### distplot(distribution plot)
check whether features are normally distributed or not

```python
df4=np.random.randn(10000)
#increasing data will approach more towards normal distribution(acc. to CLT)
```

```python
sns.distplot(df4)
#so yes, randn(rand gives uniformally distributed) always generates approximately normally distributed data
```

```python
df4=np.random.randn(100000)
sns.distplot(df4)
#68% of data is present in first std. deviation
```

### jointplot()

```python
df1.head()
```

```python
df2.head()
```

```python
sns.jointplot(x=df1['tip'], y=df2['a'], color='r')
#yes we can here connect two dataframes
```

```python
#kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }(scatter is default)
sns.jointplot(x=df1['tip'], y=df1['total_bill'], kind='hex')
```

### pairplot()

```python
sns.pairplot(df3)
#in diagonal we have histograms
```

```python
'''kind : {'scatter', 'kde', 'hist', 'reg'}
diag_kind : {'auto', 'hist', 'kde', None}'''
sns.pairplot(df1, kind='reg', diag_kind='kde', hue='day')
```
